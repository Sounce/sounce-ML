```python
import numpy as np
import cv2
import os
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
def get_gray(img):
    return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
def get_adaptive_thresh(img):
    gray = get_gray(img)
    threshold_image = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 9)
    return threshold_image
def get_thresh(img):
    gray = get_gray(img)
    gray = cv2.GaussianBlur(gray, (41, 41), 0)
    ret, threshold_image = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    return threshold_image
def get_common(img):
    return img
```

```python
def draw_images(get_img, path, l=0, r=5, cmap=plt.cm.binary):
    images_names = os.listdir(path)
    images_names.sort(key=lambda x : (len(x), x))
    for img_name in images_names[l:r]:
        image = cv2.imread(path + img_name)
        plt.figure(figsize=(3,3))
        plt.xticks([])
        plt.yticks([])
        plt.grid(False)
        img = get_img(image)
        
        mask = np.ones((224, 224))
        for i in range(35):
            mask[i] = np.zeros(224)
        # mask = np.stack((mask, ) * 3, axis=-1)
        print(img.shape)
        plt.imshow(~(cv2.bitwise_and(img, mask.astype('uint8')) * 255), cmap=cmap)
        plt.xlabel(img_name)
```

```python
datagen = ImageDataGenerator(
    horizontal_flip=True,
    vertical_flip=True,
    featurewise_std_normalization=True,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    zoom_range=0.1,
    shear_range=0.1,
)
```
